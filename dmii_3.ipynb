O sistema de recomendação utiliza dois tipos de informação, informação sobre itens (palavras-chave, categorias, etc.) e utilizadores (preferências, perfis, etc.) e informação sobre interação utilizador-item, tais como classificações, número de compras, gostos, etc. <br>

```python
def recomendacoes(users_brand_list, brand_popularity,n):
  ubl = users_brand_list.copy()
  ubl['recomendation'] = np.empty((len(ubl), 0)).tolist()
  ubl = ubl.reset_index()
  brand_popularity = brand_popularity.reset_index()
  count = 0
  teste = ubl.recomendation.copy()

  for i in range(len(ubl)):
    l = []
    brands = ubl.brand_id[i]
    for j in range(len(brand_popularity)):
      
      brand = brand_popularity.brand_id[j]
      if count < n:
        if brand not in brands:
          l.append(brand)
          count += 1
      else:
        count = 0
        break
      ubl.recomendation[i] = l
  
  return ubl



ubl = recomendacoes(users_brand_list, brand_popularity, 3)
ubl
```

```python
def merge_df(users_brand_list, users_brand_list2):
  df_merged = pd.merge(users_brand_list, users_brand_list2, on='user_id', how='left')
  #df_merged.rename(columns={'brand_id_x':'bought 161/162','brand_id_y':'bought 163'})
  
  df_merged = df_merged.rename(columns={"brand_id_x": "compras_1_184", "brand_id_y": "compras_185_231"})
  df_merged["compras_185_231"].fillna(0,inplace=True)
  return df_merged

df2 = merge_df(ubl, users_brand_list2)

df2 = df2.rename(columns={"brand_id_x": "compras_1_184", "brand_id_y": "compras_185_231"})
df2["compras_185_231"].fillna(0,inplace=True)
df2
```

```python
def create_columns(df_merge):
  df_merge['compras_recomendadas'] = np.empty((len(df_merge), 0)).tolist()
  for i in range(len(df_merge)):
    if df_merge.compras_185_231[i] != 0:
      df_merge['compras_recomendadas'][i] = list(set(df_merge.recomendation[i]) & set(df_merge.compras_185_231[i]))


  df_merge['compras_recomendadas_len'] = df_merge["compras_recomendadas"].apply(len)

  df_merge['nr_compras'] = 0
  for i in range(len(df_merge)):
    if df_merge.compras_185_231[i] != 0:
      df_merge['nr_compras'][i] = len(df_merge.compras_185_231[i])

  df_merge2 = df_merge[~df_merge.compras_recomendadas.str.len().eq(0)]

  
  df_merge['recall'] = df_merge["compras_recomendadas_len"]/df_merge["nr_compras"]  
  df_merge["recall"].fillna(0,inplace=True)
  df_merge = df_merge.drop(['index'], axis=1)

  return df_merge, df_merge2  

df_merge, df_merge2 = create_columns(df2)

df_merge
```

```python
import math

def encode_units(x):
    if x <= 0 or math.isnan(x):
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
frequent_itemsets=frequent_itemsets.sort_values(by="length",ascending=False)
frequent_itemsets
```

```python
def recomendacoes_association_rules(users_brand_list, rules, n):
  ubl = users_brand_list.copy()
  
  #ubl = ubl.loc[ubl['brand_id'].map(len) >= 0]
  ubl.sort_values('user_id')
  ubl['recomendation'] = np.empty((len(ubl), 0)).tolist()
  ubl = ubl.reset_index()
  count = 0
  for i in range(len(ubl)):
    l = []
    brands = ubl.brand_id[i]
    for j in range(len(rules)):
      
      brand = rules.antecedents[j]
      recom = rules.consequents[j]
      match = [ls for ls in brand if ls in brands]
      if count < n:
        if len(match) == len(brand):
          for n in recom:
            l.append(n)
            count += 1
      else:
        count = 0
        break
      ubl.recomendation[i] = l
  
  return ubl
asd = recomendacoes_association_rules(users_brand_list, rules, 3)
asd
#users
```
